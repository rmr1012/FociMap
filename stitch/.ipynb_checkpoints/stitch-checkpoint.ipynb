```python
import cv2
import matplotlib 
from matplotlib import pyplot as plt
import os
from huHann import huHann
import scipy
import scipy.signal as sig
import numpy as np
import pickle
```

```python
files=["6.JPG","7.JPG","8.JPG","9.JPG","10.JPG","11.JPG","12.JPG"]
focal=[6     , 12.  ,   18,    24,     36,     72,       200]
blurmaps=[]
for index, filename in enumerate(files):
    print(filename)
    im = cv2.imread(filename)
    img=im[:,:,2].astype(int)
    blurmap=huHann(img)
    blurmaps.append(blurmap)
#     plt.subplot(1,2,1)
#     plt.imshow(im)
#     plt.subplot(1,2,2)
#     plt.imshow(blurmap,cmap='gray')
#     plt.colorbar()
#     plt.savefig("phoneSet/HH"+filename[:-4]+".png")
#     plt.close()
```

```python
len(blurmaps)
```

```python
# normalize blurmaps
# 1 being on focus, 0 being out of focus
norMaps=[]
for bmap in blurmaps:
    nmap=(8-bmap)/8
    norMaps.append(nmap)
```

```python
plt.imshow(nmap)
```

```python
# bluredMaps=[]
# #kernelG=cv2.getGaussianKernel([10,10],10)

# for nnmap in norMaps:
#     blur = cv2.GaussianBlur(nnmap,(499,499),0)
#     bluredMaps.append(blur)
```

```python
# plt.imshow(blur)
```

```python
#fgenerate confidence map
confMaps=[]
#kernelG=cv2.getGaussianKernel([10,10],10)

for nnmap in norMaps:
    blur = cv2.GaussianBlur(nnmap,(99,99),0)
    blur = cv2.GaussianBlur(blur,(99,99),0)
    confMaps.append(blur)
```

```python
plt.imshow(confMaps[1])
plt.colorbar()
```

```python
flatConf=np.array(confMaps)
```

```python
flatConf.shape
```

```python
len(flatConf[0,:,0])
```

```python
oneDepth=np.ones([len(flatConf[0,:,0]),len(flatConf[0,0,:])])*999
for row in range(len(flatConf[0,:,0])):
    for col in range(len(flatConf[0,0,:])):
        corr=flatConf[:,row,col]
        maxInd=corr.argmax()
        #print(maxInd)
        oneDepth[row,col]=focal[maxInd]
```

```python
len(corr)
```

```python
oneDepth
```

```python
plt.imshow(oneDepth,norm=matplotlib.colors.LogNorm(),cmap='gray')
plt.colorbar()
```

```python
smoothDepth=cv2.blur(oneDepth,(100,100))
```

```python
plt.imshow(smoothDepth,norm=matplotlib.colors.LogNorm(),cmap='gray')
plt.colorbar()
```
